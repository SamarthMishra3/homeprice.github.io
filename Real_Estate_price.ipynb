```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1= pd.read_csv('Bengaluru_House_Data.csv')
df1.head()
```

```python
df1.groupby('area_type')['area_type'].agg('count')
```

```python
df2= df1.drop(['area_type', 'availability', 'society', 'balcony'], axis='columns')
df2.head()
```

```python
df2.isnull().sum()
```

```python
df3= df2.dropna()
df3.isnull().sum()
```

```python
df3['size'].unique()
```

```python
df3['bhk']= df3['size'].apply(lambda x: int(x.split(' ')[0]))
```

```python
df3.head()
```

```python
df3['bhk'].unique()
```

```python
df3[df3.bhk>20]
```

```python
df3.total_sqft.unique()
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
 df3[~df3['total_sqft'].apply(is_float)].head(9)
```

```python
def sqft_to_num(x):
    tokens= x.split('-')
    if len(tokens)== 2:
        return(float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df4= df3.copy()
df4['total_sqft']=df4['total_sqft'].apply(sqft_to_num)
df4.head(9)
```

```python
df5= df4.copy()
df5['price_per_sqft']= df5['price']*100000/ df5['total_sqft']
df5.head()
```

```python
len(df5.location.unique())
```

```python
df5.location= df5.location.apply(lambda x: x.strip())
loc_stats=df5.groupby('location')['location'].agg('count').sort_values(ascending=False)
loc_stats
```

```python
len(loc_stats[loc_stats<=10])
```

```python
loc_less_10=loc_stats[loc_stats<=10]
```

```python
df5.location= df5.location.apply(lambda x: 'other' if x in loc_less_10 else x)
len(df5.location.unique())
```

```python
df5[df5.total_sqft/ df5.bhk <300 ].head()
```

```python
df6=df5[~(df5.total_sqft/ df5.bhk <300)]
df6.head()
```

```python
df6.price_per_sqft.describe()
```

```python
def rem_pps_out(df):
    df_out=pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m= np.mean(subdf.price_per_sqft)
        sd= np.std(subdf.price_per_sqft)
        reduced_df= subdf[(subdf.price_per_sqft<=(m+sd)) & (subdf.price_per_sqft>(m-sd))]
        df_out= pd.concat([df_out, reduced_df], ignore_index=True)
    return df_out

df7= rem_pps_out(df6)
df7.shape
```

```python
def plot_scatter(df, location):
    bhk2=df[(df.location==location) & (df.bhk==2)]
    bhk3=df[(df.location==location) & (df.bhk==3)]
    plt.scatter(bhk2.total_sqft, bhk2.price_per_sqft, color='blue', label='2BHK', s=50)
    plt.scatter(bhk3.total_sqft, bhk3.price_per_sqft, color='green', label='3BHK',marker='+', s=50)
    plt.legend()
    
plot_scatter(df7, 'Hebbal')
```

```python
def rem_bhk_out(df):
    exclude_indices= np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats={}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk]={
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats= bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exculde_indices=np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices, axis='index')

df8= rem_bhk_out(df7)
df8.shape
```

```python
df8.bath.unique()
```

```python
df8[df8.bath>df8.bhk+2]
```

```python
df9= df8[df8.bath<df8.bhk+2]
df9.shape
```

```python
df10= df9.drop(['size', 'price_per_sqft'], axis='columns')
df10.head()
```

```python
dum= pd.get_dummies(df10.location)
```

```python
df11= pd.concat([df10, dum.drop('other', axis= 'columns')], axis='columns')
df11.head()
```

```python
df12= df11.drop('location', axis='columns')
df12.head()
```

```python
X= df12.drop('price', axis='columns')
X.head()
```

```python
y=df12.price
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
cv= ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
cross_val_score(LinearRegression(), X, y, cv=cv)
```

```python
def predict_price(location, sqft, bath, bhk):
    loc_index= np.where(X.columns==location)[0][0]
    x= np.zeros(len(X.columns))
    x[0]= sqft
    x[1]= bath
    x[2]= bhk
    if loc_index>=0:
        x[loc_index]=1
        
    return model.predict([x])[0]
```

```python
predict_price('Rajaji Nagar', 1000, 2, 2)
```
